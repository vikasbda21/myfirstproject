```python
# find the top 3 genres by total earning
df.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
df.groupby('Genre')['Gross'].sum()
```

```python
df.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with highest avg IMDB rating
df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)
```

```python
# find director with most popularity
df.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
genres = df.groupby('Genre')
genres
```

```python
genres.describe()
```

```python
genres.nunique()
```

```python
# agg method
# passing dict
genres.agg(
    {
       'Runtime':'max',
        'IMDB_Rating':'mean',
        'No_of_Votes':'mean',
        'Gross':'mean',
        'Metascore':'mean'
   }
)
```

```python
# passing list
genres.agg(['min','max','mean'])
```

```python
genres.agg(
    {
       'Runtime':['min','max'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'mean'
   }
)
```
