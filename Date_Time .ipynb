```python
# join
df1 = pd.DataFrame(
      {
          'key':['K0','K1','K2','K3','K4','K5'],
          'A':['A0','A1','A2','A3','A4','A5']
      }
)
df1
```

```python
df2 = pd.DataFrame(
      {
          'key':['K0','K1','K2','K6'],
          'B':['A0','A1','A2','A6']
      }
)
df2
```

```python
df1.join(df2,lsuffix='_caller' )
```

```python
df1.join(df2.set_index('key'),lsuffix='_caller', how='right' ,on='key' )
```

```python
df1.join(df2.set_index('key'),lsuffix='_caller', how='outer' ,on='key' )
```
